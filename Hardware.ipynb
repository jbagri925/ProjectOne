```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
# Reference the file where the CSV is located
hardware_csv_path = "Resources/refresh.csv"

# Import the data into a Pandas DataFrame
hardware_df = pd.read_csv(hardware_csv_path)
hardware_df.head()
```

```python
# Get a list of all of our columns for easy reference
hardware_df.columns
```

```python
# Extract "Serial #, "Part ID", "Part Desc.", "Within 5 years- Supported", "Beyond 5 Years", "Within 5 years - Unsupported"
warranty_hardware_df = hardware_df.loc[:, ["Serial #", "Part ID", "Part Desc.","Wrty Start Date","5 Year End Date.",
                                    "Within 5 years- Supported", "Beyond 5 Years", "Within 5 years - Unsupported"]]
warranty_hardware_df.head()
```

```python
# attribute to return the year of the datetime 
warranty_hardware_df['5 Year End Date.'] =  pd.to_datetime(warranty_hardware_df['5 Year End Date.'],infer_datetime_format=True)
warranty_hardware_df.head()
```

```python
#Convert date to year
warranty_hardware_df['5 Year End Date.']=warranty_hardware_df['5 Year End Date.'].dt.year

# check the datatypes
warranty_hardware_df.dtypes
```

```python
#Servers with warranty end date after 2012
warranty_hardware_df.loc[warranty_hardware_df['5 Year End Date.'] > 2012]
warranty_hardware_df.head()
```

```python
# Count number of servers that expire by year. 
df2 = warranty_hardware_df.groupby(['Part Desc.', '5 Year End Date.']).size().reset_index(name='count')
df2[['5 Year End Date.']] = df2[['5 Year End Date.']].astype(int)
df2
```

```python
# Graph displays how many servers of each type expire in 2017,2018 and 2019
warranty_hardware_df.groupby(['5 Year End Date.','Part ID']).size().unstack().plot(kind='bar',stacked=True,grid=True, title="Hardware Warranty")
plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```
